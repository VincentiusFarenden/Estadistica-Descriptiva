e) Si se sabe que la persona escogida al azar tiene contrato a plazo fijo, ¬øcu√°l es la probabilidad que sea un profesional?

```python
Tabla4 = pd.crosstab( df_SUELDOS['Estamento'], df_SUELDOS['Tipo de contrato'],
                     normalize='columns', margins= True)*100
round(Tabla4, 2)
#normalize='columns' hace que cada celda es la raz√≥n del respectivo valor con respecto al total de la columna
```

Respuesta: Observamos que si la persona tiene contratoa plazo fijo, la probabilidad de ser profesional es de un 11.06%
