```python
#EJEMPLO 1
#TABLA FRECUENCIA a partir de serie de datos
#Variable: Dispositivos conectados (cuantitativa discreta)

import pandas as pd

conectados = pd.Series([3, 2, 4, 3, 2, 1, 4, 3, 3, 5,
                2, 1, 3, 4, 2, 2, 3, 4, 1, 5,
                3, 3, 2, 1, 3, 4, 5, 2, 3, 3])
tamaño = 30
frecuencias = conectados.value_counts().sort_index()
tabla_frecuencias_ejemplo_1 = pd.DataFrame({
    'Frec. absoluta': frecuencias,
    'Frec. relativa (%)': (frecuencias/tamaño)*100,
    'Frec. abs. acumulada': frecuencias.cumsum(),
    'Frec. rel. acumulada (%)': (frecuencias.cumsum()/tamaño)*100})
tabla_frecuencias_ejemplo_1
round(tabla_frecuencias_ejemplo_1,1)
```

```python
#EJEMPLO 2
#TABLA FRECUENCIA dada la frec absoluta
#Variable: Preferencia musical (cualitativa nominal)

import pandas as pd

Tabla = pd.Series([4, 3, 6, 4, 8],
 ['Rock', 'Heavy Metal', 'Pop', 'Reggaetón', 'Otros'])
tamaño = 25
tabla_frecuencias_ejemplo_2 = pd.DataFrame({
    'Frec. absoluta': Tabla,
    'Frec. relativa (%)': (Tabla/tamaño)*100})
tabla_frecuencias_ejemplo_2
```

```python
#EJEMPLO 3
#TABLA FRECUENCIA a partir de columna excel
#Variable: Velocidad internet (cuantitativa continua)

import pandas as pd

df_resumen = pd.read_excel('BBDD (Velocidad).xlsx')
intervalos = 8
# Vamos a agregar una nueva columna, con la función cut()
df_resumen['Intervalos'] = pd.cut(df_resumen['Velocidad (Mbps)'], bins = intervalos, include_lowest=True)
# Frecuencias
f = df_resumen.groupby('Intervalos').size()
F = f.cumsum()
h = (f/40)*100
H = h.cumsum()
# Armar el DataFrame
tabla_frecuencias_ejemplo_3 = pd.DataFrame({
    'Frecuencia absoluta': f,
    'Frecuencia relativa (%)': h,
    'Frecuencia absoluta acumulada': F,
    'Frecuencia relativa acumulada (%)': H})
tabla_frecuencias_ejemplo_3
```

```python
#EJEMPLO 4
#GRAFICO CIRCULAR, a partir de TABLA FRECUENCIA dada
#Variable: Preferencia musical (cualitativa nominal)
#Se grafica la frecuencia relativa

import matplotlib.pyplot as plt

etiquetas_g_circular = tabla_frecuencias_ejemplo_2.index
f_h_circular = tabla_frecuencias_ejemplo_2['Frec. relativa (%)']
fig, ax = plt.subplots(figsize=(6, 4))
circular_musical = ax.pie(f_h_circular, labels=etiquetas_g_circular, autopct='%.1f%%')
ax.set_title('Distribución de Géneros Musicales')
plt.show()
```

```python
#EJEMPLO 5
#GRAFICO BARRA, a partir de TABLA FRECUENCIA dada
#Variable: Preferencia musical (cualitativa nominal)
#Se grafica la frecuencia absoluta

import matplotlib.pyplot as plt

etiquetas_g_columna = tabla_frecuencias_ejemplo_2.index
f_a_columna = tabla_frecuencias_ejemplo_2['Frec. absoluta']
fig, ax = plt.subplots(figsize=(6, 4))
columnas_musical = ax.bar(etiquetas_g_columna, f_a_columna, color = 'skyblue', edgecolor = 'black')
ax.bar_label(columnas_musical)
ax.set_title('Distribución de Géneros Musicales')
ax.set_xlabel('Géneros musicales')
ax.set_ylabel('Cantidad de alumnos')
plt.show()
```

```python
#EJEMPLO 3
#HISTOGRAMA, a partir de TABLA FRECUENCIA a partir de columna excel
#Variable: Velocidad internet (cuantitativa continua)
#Se grafica la frecuencia absoluta

import matplotlib.pyplot as plt

etiquetas_tiempo_histograma = tabla_frecuencias_ejemplo_3.index
f_a_histograma = tabla_frecuencias_ejemplo_3['Frecuencia absoluta']
fig, ax = plt.subplots(figsize=(6, 4))
histograma_tiempo = ax.bar(etiquetas_tiempo_histograma.astype(str), f_a_histograma,
                           color = 'violet', edgecolor = 'black', width=1)
ax.bar_label(histograma_tiempo)
ax.set_title('Distribución de la velocidad de conexión')
ax.set_xlabel('Velocidad de conexión (Mbps)')
ax.set_ylabel('Cantidad de registros')
plt.xticks(rotation=45, ha='right')
plt.show()
```
