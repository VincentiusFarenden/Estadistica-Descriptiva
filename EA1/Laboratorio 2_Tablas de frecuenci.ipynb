3. Crea una tabla de frecuencias puntuales y acumuladas utilizando la variable *Puntaje de felicidad*. Considera crear la tabla de frecuencias con 8 intervalos.

```python
df_3 = pd.read_csv('Reporte_felicidad_2024.csv')
intervalos = 8
# Vamos a agregar una nueva columna, con la funci√≥n cut()
df_3['Intervalos'] = pd.cut(df_3['Puntaje de felicidad'], bins = intervalos, include_lowest=True)
# Frecuencias
f = df_3.groupby('Intervalos').size()
F = f.cumsum()
h = ((f/140)*100).round(1)
H = h.cumsum()
# Armar el DataFrame
tabla_frecuencias_ejemplo_3 = pd.DataFrame({
    'Frecuencia absoluta fi': f,
    'Frecuencia relativa hi (%)': h,
    'Frecuencia absoluta acumulada Fi': F,
    'Frecuencia relativa acumulada Hi (%)': H})
tabla_frecuencias_ejemplo_3
```

```python
df_4 = pd.read_excel('Llamadas.xlsx')

intervalos = 9
df_4['Intervalos'] = pd.cut(df_4['Tiempo (en minutos)'], bins=intervalos, include_lowest=True)

# Frecuencias
f = df_4.groupby('Intervalos').size()
F = f.cumsum()
h = ((f / 250) * 100).round(1)  # Total de observaciones es 250, en porcentaje, redondeado a 1 decimal
H = h.cumsum()

# Armar el DataFrame
tabla_frecuencias_ejemplo_4 = pd.DataFrame({
    'Frecuencia absoluta fi': f,
    'Frecuencia relativa hi (%)': h,
    'Frecuencia absoluta acumulada Fi': F,
    'Frecuencia relativa acumulada Hi (%)': H
})

# Mostrar la tabla
tabla_frecuencias_ejemplo_4
```

1. Construye una tabla de distribuci√≥n de frecuencias (absolutas y porcentuales incluyendo las acumuladas) que resuma los puntajes obtenidos en la prueba SIMCE de Matem√°tica. Considera 9 intervalos para la construcci√≥n de la tabla.

$\hspace{1cm}$*Observaciones: Se debe escoger la variable prom_mate8b_com*

```python
import pandas as pd

# Definir intervalos manualmente
bins = [200, 213, 226, 239, 252, 265, 278, 291, 304, 317]
df_5['Intervalos'] = pd.cut(df_5['prom_mate8b_com'], bins=bins, include_lowest=True, 
                           labels=[f'[{bins[i]}, {bins[i+1]})' for i in range(len(bins)-1)])

# Calcular frecuencias
f = df_5.groupby('Intervalos').size()
F = f.cumsum()
h = ((f / 344) * 100).round(1)  # Total de comunas es 344, porcentaje redondeado a 1 decimal
H = h.cumsum()

# Crear la tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Frecuencia absoluta (f_i)': f,
    'Frecuencia relativa (h_i %)': h,
    'Frecuencia absoluta acumulada (F_i)': F,
    'Frecuencia relativa acumulada (H_i %)': H
})

# Mostrar la tabla
tabla_frecuencias
```
