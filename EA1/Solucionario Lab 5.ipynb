3. Calcula el rango de la variable *Pago base*, 쯘ste valor representa la realidad de todos los trabajadores?

```python
#Calculamos el m치ximo y m칤nimo del sueldo base
maximo_pago = df_1['Pago base'].max()
minimo_pago = df_1['Pago base'].min()

#Calculamos el rango restando el m치ximo con el m칤nimo
rango = maximo_pago - minimo_pago
print(f'Rango: {rango} d칩lares.')
```

La desviaci칩n est치ndar es de 31.122 d칩lares y la media es de 77.087 d칩lares, por lo que el coeficiente de variaci칩n es aproximadamente 40,4%, esto significa que la dispersi칩n es alta con respecto promedio.

Aunque la mediana (74.711,5 d칩lares) es cercana a la media, la elevada dispersi칩n sugiere que existen diferencias importantes entre los sueldos de los trabajadores.
```python
#Creamos una variable con los cuatro cargos con mayor presencia
cargos_frecuentes = ['Enfermera Especial', 'L칤der De Recreaci칩n',
                     'Auxiliar De Servicio P칰blico - Obras P칰blicas',
                     'Becario De Servicio P칰blico']

#Creamos un nuevo DF con la informaci칩n de los cuatro cargos con mayor frecuencia
#Para esto filtramos el DF original con la funci칩n isin()
df3_filtrado = df_3[df_3['T칤tulo profesional'].isin(cargos_frecuentes)]

#Primeras cinco filas
df3_filtrado.head()
```
