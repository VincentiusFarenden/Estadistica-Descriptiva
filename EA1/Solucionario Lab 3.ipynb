```python
df_1 = pd.read_excel('BDActividad_1_2_3L3.xlsx')

#Crearemos una nueva columna llamada Intervalos_pib
df_1['Intervalos_edad'] = pd.cut(df_1['Edad'], bins = 10, include_lowest= True, precision = 0)

#Calcularemos la frecuencia absoluta de la variable Intervalos_pib
cuenta_edad = df_1.groupby('Intervalos_edad', observed = True).size()

tabla_3 = pd.DataFrame({
    'Frecuencia absoluta': cuenta_edad,
    'Frecuencia absoluta acumulada': cuenta_edad.cumsum(),
    'Frecuencia relativa (%)': cuenta_edad/100000 * 100,
    'Frecuencia relativa acumulada (%)': (cuenta_edad/100000 * 100).cumsum()
})
round(tabla_3, 1)
```

9. Calcula el porcentaje de personas que usa internet en cada pa칤s y clasif칤calos en 9 intervalos. Luego, construye un histograma que muestre la distribuci칩n de pa칤ses seg칰n ese porcentaje. 쯈u칠 observas?

```python
#Crearemos una tabla con el c치lculo de la proporci칩n relativa de usuarios
#de internet por pa칤s
tabla_7 = pd.DataFrame({
    'Proporci칩n por pa칤s (%)': df_2['USUARIOS INTERNET(marzo 2017)'] /
    df_2['POBLACI칍N (a침o 2017)'] * 100
})

#Crearemos una nueva columna llamada Intervalos_internet
tabla_7['Intervalos_internet'] = pd.cut(tabla_7['Proporci칩n por pa칤s (%)'],
                                        bins = 9, include_lowest= True, precision = 1)

#Calcularemos la frecuencia absoluta de la variable Intervalos_internet
cuenta_intervalo = tabla_7.groupby('Intervalos_internet', observed = True).size()


#Ahora crearemos el histograma
#La frecuencia absoluta (altura de las barras) la sacaremos de la variable
#cuenta_intervalo
categorias_intervalo = cuenta_intervalo.index


#Gr치fico
fig, ax = plt.subplots(figsize = (7,5))
histograma_internet= ax.bar(categorias_intervalo.astype(str), cuenta_intervalo,
                            color = 'tomato', edgecolor = 'black', width = 1)

#Personalizaci칩n
ax.set_title('Distribuci칩n de uso de internet por pa칤s')
ax.set_xlabel('Porcentaje de usuarios')
ax.set_ylabel('Cantidad de pa칤ses')
ax.bar_label(histograma_internet)
plt.xticks(rotation=45, ha = 'right') #Rota en 45춿 las categor칤as para evitar la superposici칩n
plt.show()
```

Esto evidencia una fuerte brecha digital en el a침o 2017, donde una parte importante de los pa칤ses ten칤a a칰n un acceso muy limitado a internet.

*Observaci칩n: Considerar que el valor m칤nimo del primer intervalo aparece como -0,2% debido al uso del argumento include_lowest=True en la funci칩n cut() de Pandas. Este argumento asegura que el valor m치s bajo de la variable (en este caso, un pa칤s con 0% de uso de internet) sea incluido en el primer intervalo. Para lograr esto, la funci칩n extiende levemente el l칤mite inferior del primer intervalo hacia la izquierda en la recta num칠rica.*
