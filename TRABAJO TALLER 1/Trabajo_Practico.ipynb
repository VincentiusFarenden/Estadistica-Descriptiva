## **Tabla de frecuencias Variable EDAD**

```python

import pandas as pd

# Tomar la columna correcta
EDAD = df_1['EDAD']  # Usando exactamente el nombre de la columna
tamaño = len(EDAD)

# Frecuencias
frecuencias = EDAD.value_counts().sort_index()

# Tabla de frecuencias
tabla_frecuencias_EDAD = pd.DataFrame({
    'Frec. absoluta fi': frecuencias,
    'Frec. relativa hi (%)': (frecuencias / tamaño) * 100,
    'Frec. abs. acumulada Fi': frecuencias.cumsum(),
    'Frec. rel. acumulada Hi(%)': (frecuencias.cumsum() / tamaño) * 100
})

# Redondear a 1 decimal
tabla_frecuencias_EDAD = round(tabla_frecuencias_EDAD, 1)

tabla_frecuencias_EDAD
```

## **Histograma de Variable EDAD**



```python
import matplotlib.pyplot as plt

# Tomamos el índice (EDADes) y las frecuencias absolutas
etiquetas_EDAD = tabla_frecuencias_EDAD.index
f_a_EDAD = tabla_frecuencias_EDAD['Frec. absoluta fi']

fig, ax = plt.subplots(figsize=(22, 10))
histograma_EDAD = ax.bar(etiquetas_EDAD.astype(str), f_a_EDAD,
                         color='violet', edgecolor='black', width=0.9)

# Etiquetas arriba de cada barra
ax.bar_label(histograma_EDAD)

ax.set_title('Distribución de la EDAD de matriculados region del Maule')
ax.set_xlabel('EDAD (años)')
ax.set_ylabel('Cantidad de Matriculados')

plt.xticks(rotation=45, ha='right')
plt.show()
```

## **Tabla de frecuencias de Valor Arancel (Pesos)**

```python
import numpy as np
import pandas as pd

VALOR_ARANCEL = df_1['VALOR ARANCEL (PESOS)']

# Paso 1: Rango
R = VALOR_ARANCEL.max() - VALOR_ARANCEL.min()

# Paso 2: Número de clases (Sturges)
n = len(VALOR_ARANCEL)
k = int(1 + 3.322 * np.log10(n))

# Paso 3: Amplitud de clase
A = int(np.ceil(R / k))

print("Rango (R):", R)
print("Número de clases (k):", k)
print("Amplitud de clase (A):", A)

# Paso 4: Construcción de intervalos
bins = np.arange(VALOR_ARANCEL.min(), VALOR_ARANCEL.max() + A, A)
intervalos = pd.cut(VALOR_ARANCEL, bins=bins)

# Paso 5: Tabla de frecuencias
frecuencias = intervalos.value_counts().sort_index()

tabla_frecuencias_ARANCEL = pd.DataFrame({
    'Intervalo': frecuencias.index.astype(str),
    'Frec. absoluta fi': frecuencias.values,
    'Frec. relativa hi (%)': (frecuencias.values / n) * 100,
    'Frec. abs. acumulada Fi': np.cumsum(frecuencias.values),
    'Frec. rel. acumulada Hi (%)': np.cumsum(frecuencias.values) / n * 100
})

# Redondear decimales a 1
tabla_frecuencias_ARANCEL = tabla_frecuencias_ARANCEL.round(1)

tabla_frecuencias_ARANCEL
```

## **Histograma de Valor Arancel (Pesos)**




```python
import matplotlib.pyplot as plt

# Usar directamente frecuencias y su índice
etiquetas_arancel = frecuencias.index.astype(str)
f_a_arancel = frecuencias.values

fig, ax = plt.subplots(figsize=(12, 6))
barras = ax.bar(etiquetas_arancel, f_a_arancel,
                color='violet', edgecolor='black', width=0.9)

# Etiquetas encima de cada barra
ax.bar_label(barras)

# Títulos y ejes
ax.set_title('Distribución del Valor de Arancel (Pesos)', fontsize=14)
ax.set_xlabel('Intervalos de arancel (pesos)')
ax.set_ylabel('Cantidad de estudiantes')

plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()

```

## **Tabla de frecuencias de Duración del Plan de Estudio (Semestres)**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Tomar la columna
DURACION = df_1['DURACION PLAN DE ESTUDIO (SEMESTRES)']

# Tamaño
n = len(DURACION)

# Frecuencias absolutas y relativas
frecuencias = DURACION.value_counts().sort_index()
frecuencia_relativa = (frecuencias / n * 100).round(1)
frecuencia_acumulada = frecuencias.cumsum()
frecuencia_relativa_acum = frecuencia_relativa.cumsum()

# Crear tabla de frecuencias
tabla_frecuencias_DURACION = pd.DataFrame({
    'Frec. absoluta fi': frecuencias,
    'Frec. relativa hi (%)': frecuencia_relativa,
    'Frec. abs. acumulada Fi': frecuencia_acumulada,
    'Frec. rel. acumulada Hi (%)': frecuencia_relativa_acum
})

tabla_frecuencias_DURACION
```

## **Histograma de Duración del Plan de Estudio (Semestres)**

```python
# Gráfico de barras con etiquetas
etiquetas = tabla_frecuencias_DURACION.index.astype(str)
f_a = tabla_frecuencias_DURACION['Frec. absoluta fi']

fig, ax = plt.subplots(figsize=(8, 5))
barras = ax.bar(etiquetas, f_a, color='violet', edgecolor='black', width=0.7)

# Agregar etiquetas encima de cada barra
ax.bar_label(barras)

# Títulos y ejes
ax.set_title('Distribución de la Duración del Plan de Estudio (Semestres)', fontsize=14)
ax.set_xlabel('Semestres')
ax.set_ylabel('Cantidad de matriculados')

plt.show()
```
