## **Tabla de frecuencias Variable EDAD**

```python
import numpy as np
import pandas as pd


EDAD = df_1['EDAD']  # Usando exactamente el nombre de la columna

# Paso 1: Rango
R = EDAD.max() - EDAD.min()

# Paso 2: Número de clases (Sturges)
n = len(EDAD)
k = int(1 + 3.322 * np.log10(n))

# Paso 3: Amplitud de clase
A = int(np.ceil(R / k))

print("Rango (R):", R)
print("Número de clases (k):", k)
print("Amplitud de clase (A):", A)

# Paso 4: Construcción de intervalos
bins = np.arange(EDAD.min(), EDAD.max() + A, A)
intervalos_EDAD = pd.cut(EDAD, bins=bins)

# Paso 5: Tabla de frecuencias
frecuencias_EDAD = intervalos_EDAD.value_counts().sort_index()

tabla_frecuencias_EDAD = pd.DataFrame({
    'Intervalo': frecuencias_EDAD.index.astype(str),
    'Frec. absoluta fi': frecuencias_EDAD.values,
    'Frec. relativa hi (%)': (frecuencias_EDAD.values / n) * 100,
    'Frec. abs. acumulada Fi': np.cumsum(frecuencias_EDAD.values),
    'Frec. rel. acumulada Hi (%)': np.cumsum(frecuencias_EDAD.values) / n * 100
})

# Redondear decimales a 1
tabla_frecuencias_EDAD = tabla_frecuencias_EDAD.round(1)

tabla_frecuencias_EDAD
```

2. Se requiere saber cual es el promedio de los costes de arancel de los establecimientos educativos, de los cuales para el siguente analisis se tendran en cuenta: Universidades privadas, Institutos Profesionales Y Centros de Formacion Tecnica.
De esta manera podremos estar al tanto de la situacion economica de los distintos establecimientos.




```python
import pandas as pd

df = pd.read_excel('09_MATRICULAS_ED_SUPERIOR_MAULE_2021.xlsx')

#Se toma los datos de los distintos establecimientos eduacionales para sacarle el promedio del arancel
promedio_universidades = df[df['TIPO DE INSTITUCION'] == 'Universidades Privadas']['VALOR ARANCEL (PESOS)'].mean()
promedio_institutos = df[df['TIPO DE INSTITUCION'] == 'Institutos Profesionales']['VALOR ARANCEL (PESOS)'].mean()
promedio_centro_ed_tecnica = df[df['TIPO DE INSTITUCION'] == 'Centros de Formacion Tecnica']['VALOR ARANCEL (PESOS)'].mean()

#Se crea una tabla que muestre los promedios de los establecimientos
tabla_promedios = pd.DataFrame({
    'Tipo de Institución': ['Universidades Privadas', 'Institutos Profesionales', 'Centros de Formacion Tecnica'],
    'Promedio de Arancel ($)': [promedio_universidades, promedio_institutos, promedio_centro_ed_tecnica]
}).round(2)

tabla_promedios
```

De acuerdo a los datos entregados anteriormente se ve que los centros de formacion tecnica e institutos tienen un promedio de valor de arancel muy similar, sin embargo en las universidades privadas hay una diferencia de casi el doble a comparacion a de los otros establecimientos, lo que quiere decir que estudiantes tendran una mayor carga financiera si deciden optar por las universidades privadas en lugar de un instituto profesional o Centro de formacion tecnica

Estos datos quieren decir que en ambas hay una gran variedad de costes entre las universidades cruch y privadas, sin embargo las universidades cruch tiene costes mas uniformes en comparacion que las universidades privadas. Pese a ello la varianza entre ambas sea mas o menos similar.

4. Para tener un mejor analisis, se considerara usar la desviacion estandar para determinar que costes de aranceles es mas consistente, comparando entre las universidades cruch, y universidades privadas.

```python
import pandas as pd

df = pd.read_excel('09_MATRICULAS_ED_SUPERIOR_MAULE_2021.xlsx')

#Se unen los datos entre universidades privadas o universidades cruch, y luego se hace la union con el valor de arancel y se calcula la desviacion estandar
Desv_un_privadas = df[df['TIPO DE INSTITUCION'] == 'Universidades Privadas']['VALOR ARANCEL (PESOS)'].std()
Desv_un_CRUCH  =df[df['TIPO DE INSTITUCION'] == 'Universidades CRUCH']['VALOR ARANCEL (PESOS)'].std()

tabla_desv_estandar = pd.DataFrame({
    'Tipo de Universidad': ['Universidades Privadas', 'Universidades del Cruch'],
    'Desviación Estándar de Arancel ($)': [Desv_un_privadas, Desv_un_CRUCH]
}).round(2)

tabla_desv_estandar
```

Con esto podemos corroborar que las unverisdades privadas tienen costes mas dispersos y menos consistentes a comparacion de las universidades cruch.
Aunque la desviacion estandar sea similar entre universidades privadas y universidades cruch, siendo una diferencia de aproximadamente 10.000.

5. Se necesita comparar el costo de las carreras profesionales y el costo de las carreras tecnicas. Es por ello que se usaran diferentes que nos ayuden a corroborar la variedad de costes de los diferentes tipos de carreras

```python
import pandas as pd

df = pd.read_excel('09_MATRICULAS_ED_SUPERIOR_MAULE_2021.xlsx')

#Se hace union del costo de arancel y la matricula y se guarda en COSTO TOTAL
df['COSTO TOTAL'] = df['VALOR MATRICULA (PESOS)'] + df['VALOR ARANCEL (PESOS)']

#Se calcula el promedio del costo total de las carreras tecnico y profesional
prom_profesional = df[df['NIVEL CARRERA'] == 'Carreras Profesionales']['COSTO TOTAL'].mean()
prom_tecnico = df[df['NIVEL CARRERA'] == 'Carreras Tecnicas']['COSTO TOTAL'].mean()

#Se calcula la desviacion estandar de las carreras tecnico y profesional
desv_profesional = df[df['NIVEL CARRERA'] == 'Carreras Profesionales']['COSTO TOTAL'].std()
desv_tecnico = df[df['NIVEL CARRERA'] == 'Carreras Tecnicas']['COSTO TOTAL'].std()

#Se calcula el coeficiente de variacion de las carreras tecnico y profesional
cof_profesional = round((desv_profesional / prom_profesional) * 100,2)
cof_tecnico = round((desv_tecnico / prom_tecnico) * 100,2)


tabla_coeficientes = pd.DataFrame({
    'Tipo de Carrera': ['Profesionales', 'Técnicas'],
    'Coeficiente de Variación (%)': [cof_profesional, cof_tecnico]
}).round(2)


tabla_coeficientes
```

De los datos se puede ver que hubo un 31.68% de coeficiente de variacion en el caso de las carreras profesionales, mientras que en las carreras tecnicas hubo un 15.05% de coeficiende de variacion.
Es decir que las carreras profesionales le saca un coeficiente de variacion mas del doble que a las carreras tecnicas, por ende las carreras profesionales, cosa que en las tecnicas demuestra ser mas consistente y mas homogeneo en sus numeros, y por ende una menor dispersion.
