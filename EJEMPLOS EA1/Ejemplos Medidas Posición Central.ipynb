```python
import pandas as pd

df_SUELDOS = pd.read_excel('BBDD (Sueldo) S2_RA2.xlsx')
df_SUELDOS
```

```python
#Cálculo Media o promedio de la variable "Sueldo mensual"

promedio_sueldo = round(df_SUELDOS['Sueldo mensual\n empresa (m$)'].mean(),3)
print("El sueldo promedio de los trabajadores es ", promedio_sueldo, "miles de pesos")
promedio_sueldo_pesos = promedio_sueldo*1000
print("El sueldo promedio de los trabajadores es ", promedio_sueldo_pesos, "pesos")
```

```python
#Cálculo Media o promedio de la variable "Gasto mensual"

promedio_gasto = round(df_SUELDOS['Gasto Mensual (M$)'].mean(),3)
print("El gasto promedio de los trabajadores es ", promedio_gasto, "miles de pesos")
promedio_gasto_pesos = promedio_gasto*1000
print("El sueldo promedio de los trabajadores es ", promedio_gasto_pesos, "pesos")
```

```python
#Crear un dataframe que muestre promedios de sueldo por tipo de contrato

df_PROM_SUELDOS_TIPO = round(df_SUELDOS.groupby('Tipo de contrato')['Sueldo mensual\n empresa (m$)'].mean(),3)
df_PROM_SUELDOS_TIPO
```

```python
#Crear un dataframe que muestre promedios de sueldo por estamento
df_PROM_SUELDOS_EST = round(df_SUELDOS.groupby('Estamento')['Sueldo mensual\n empresa (m$)'].mean(),3)
df_PROM_SUELDOS_EST
```

```python
#Cálculo Moda de la variable "Sueldo mensual"
moda_sueldo = round(df_SUELDOS['Sueldo mensual\n empresa (m$)'].mode(),3)
print("El sueldo mas frecuente de los trabajadores es ", moda_sueldo, "miles de pesos")
moda_sueldo_pesos = moda_sueldo*1000
print("El sueldo mas frecuente de los trabajadores es ", moda_sueldo_pesos, "pesos")
```

```python
#Cálculo Mediana de la variable "Gasto mensual"

mediana_gasto = round(df_SUELDOS['Gasto Mensual (M$)'].median(),3)
print("El 50% de los trabajadores gasta a lo más ", mediana_gasto, "miles de pesos")
mediana_gasto_pesos = mediana_gasto*1000
print("El 50% de los trabajadores gasta a lo más ", mediana_gasto_pesos, "pesos")
```

```python
#Crear un dataframe (bivariado) que muestre promedios de sueldo por estamento y tipo contrato

TABLA = round(pd.crosstab( df_SUELDOS['Estamento'], df_SUELDOS['Tipo de contrato'],
                    df_SUELDOS['Sueldo mensual\n empresa (m$)'], aggfunc='mean'),3)
TABLA
```
