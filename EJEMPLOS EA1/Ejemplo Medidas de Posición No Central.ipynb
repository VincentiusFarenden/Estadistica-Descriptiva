El departamento de Gestión de Personas de la empresa Del Sur S.A. plantea como estrategia entregar paquetes de seguros de cesantía a los trabajadores en función de sus sueldos mensuales, sin importar el tipo de contrato. La base de datos registra los sueldos mensuales de cada trabajador en miles de pesos (M$), el tipo de contrato, el estamento al cual pertenece y el gasto mensual en miles de pesos (M$) de un determinado mes. 

```python
# Importar librerías
import pandas as pd

df_SUELDOS = pd.read_excel('BBDD_SueldoGastoRA3.xlsx')
print(df_SUELDOS.columns) # Identificar correctamente los nombres de las columnas
```

a) Se desea entregar un bono de transporte para aquellos trabajadores que pertenecen al 30% de los menores sueldos. ¿Cuál es el sueldo más alto que puede tener un trabajador para pertenecer a este grupo?

```python
#Calculamos el percentil 30
percentil_30 = round (df_SUELDOS['Sueldo mensual\n empresa (m$)'].quantile(0.3)*1000,0)
print("El sueldo máximo para permanecer al 30% con menores ingresos es ", percentil_30, "pesos")
```

b) El departamento de Gestión de Personas quiere hacer un ajuste en los sueldos de los trabajadores. Para ello quiere saber cuál es el sueldo mínimo del 15% de los sueldos más altos. ¿Cuál es el sueldo mínimo de este grupo?

```python
#Calcular el percentil 85
percentil_85 = round(df_SUELDOS['Sueldo mensual\n empresa (m$)'].quantile(0.85)*1000,0)
print("El sueldo mínimo para pertenecer al 15% con mayores ingresos es ", percentil_85, "pesos")
```

c) Analice el gasto mensual de los trabajadores utilizando los cuartiles.

```python
#Calculamos cada uno de los tres cuartiles del Gasto
C_1 = round(df_SUELDOS['Gasto Mensual (M$)'].quantile(0.25)*1000,0)
C_2 = round(df_SUELDOS['Gasto Mensual (M$)'].quantile(0.5)*1000,0)
C_3 = round(df_SUELDOS['Gasto Mensual (M$)'].quantile(0.75)*1000,0)

#Calculamos los valores extremos del Gasto
mini = round(df_SUELDOS['Gasto Mensual (M$)'].min()*1000,0)
maxi = round(df_SUELDOS['Gasto Mensual (M$)'].max()*1000,0)

print("El gasto mínimo es ", mini, "pesos")
print("El primer cuartil del gasto es ", C_1, "pesos")
print("El segundo cuartil del gasto es ", C_2, "pesos")
print("El tercer cuartil del gasto es ", C_3, "pesos")
print("El gasto máximo es ", maxi, "pesos")
```
