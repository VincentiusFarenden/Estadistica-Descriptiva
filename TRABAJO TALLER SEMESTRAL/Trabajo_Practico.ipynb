## **Tabla de frecuencias Variable EDAD**

```python
import numpy as np
import pandas as pd


EDAD = df_1['EDAD']  # Usando exactamente el nombre de la columna

# Paso 1: Rango
R = EDAD.max() - EDAD.min()

# Paso 2: Número de clases (Sturges)
n = len(EDAD)
k = int(1 + 3.322 * np.log10(n))

# Paso 3: Amplitud de clase
A = int(np.ceil(R / k))

print("Rango (R):", R)
print("Número de clases (k):", k)
print("Amplitud de clase (A):", A)

# Paso 4: Construcción de intervalos
bins = np.arange(EDAD.min(), EDAD.max() + A, A)
intervalos_EDAD = pd.cut(EDAD, bins=bins)

# Paso 5: Tabla de frecuencias
frecuencias_EDAD = intervalos_EDAD.value_counts().sort_index()

tabla_frecuencias_EDAD = pd.DataFrame({
    'Intervalo': frecuencias_EDAD.index.astype(str),
    'Frec. absoluta fi': frecuencias_EDAD.values,
    'Frec. relativa hi (%)': (frecuencias_EDAD.values / n) * 100,
    'Frec. abs. acumulada Fi': np.cumsum(frecuencias_EDAD.values),
    'Frec. rel. acumulada Hi (%)': np.cumsum(frecuencias_EDAD.values) / n * 100
})

# Redondear decimales a 1
tabla_frecuencias_EDAD = tabla_frecuencias_EDAD.round(1)

tabla_frecuencias_EDAD
```

```python
import pandas as pd

df = pd.read_excel('09_MATRICULAS_ED_SUPERIOR_MAULE_2021.xlsx')

#Se hace union del costo de arancel y la matricula y se guarda en COSTO TOTAL
df['COSTO TOTAL'] = df['VALOR MATRICULA (PESOS)'] + df['VALOR ARANCEL (PESOS)']

#Se calcula el promedio del costo total de todos los niveles de carrera
prom_profesional = df[df['NIVEL CARRERA'] == 'Carreras Profesionales']['COSTO TOTAL'].mean()
prom_tecnico = df[df['NIVEL CARRERA'] == 'Carreras Tecnicas']['COSTO TOTAL'].mean()
prom_magister = df[df['NIVEL CARRERA'] == 'Magister']['COSTO TOTAL'].mean()
prom_doctorado = df[df['NIVEL CARRERA'] == 'Doctorado']['COSTO TOTAL'].mean()
prom_postitulo = df[df['NIVEL CARRERA'] == 'Postitulo']['COSTO TOTAL'].mean()

#Se calcula la desviacion estandar de las carreras tecnico y profesional
desv_profesional = df[df['NIVEL CARRERA'] == 'Carreras Profesionales']['COSTO TOTAL'].std()
desv_tecnico = df[df['NIVEL CARRERA'] == 'Carreras Tecnicas']['COSTO TOTAL'].std()
desv_magister = df[df['NIVEL CARRERA'] == 'Magister']['COSTO TOTAL'].std()
desv_doctorado = df[df['NIVEL CARRERA'] == 'Doctorado']['COSTO TOTAL'].std()
desv_postitulo = df[df['NIVEL CARRERA'] == 'Postitulo']['COSTO TOTAL'].std()

#Se calcula el coeficiente de variacion de las carreras tecnico y profesional
cof_profesional = round((desv_profesional / prom_profesional) * 100,2)
cof_tecnico = round((desv_tecnico / prom_tecnico) * 100,2)
cof_magister = round((desv_magister / prom_magister) * 100,2)
cof_doctorado = round((desv_doctorado / prom_doctorado) * 100,2)
cof_postitulo = round((desv_postitulo / prom_postitulo) * 100,2)


tabla_coeficientes = pd.DataFrame({
    'Tipo de Carrera': ['Profesionales', 'Técnicas','Magister','Doctorado','Postitulo'],
    'Coeficiente de Variación (%)': [cof_profesional, cof_tecnico, cof_magister, cof_doctorado, cof_postitulo]
}).round(2)


tabla_coeficientes
```

```python
# Importar librerías y Base de Datos

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

df_1 = pd.read_excel('09_MATRICULAS_ED_SUPERIOR_MAULE_2021.xlsx')
df_1.head()
```

### 1. Correlación entre matrícula y arancel

### 1.1 Análisis general

```python
# Covarianza
covarianza = df_1['VALOR MATRICULA (PESOS)'].cov(df_1['VALOR ARANCEL (PESOS)'])

# Coeficiente de correlación lineal de Pearson
coef_correlacion = df_1['VALOR MATRICULA (PESOS)'].corr(df_1['VALOR ARANCEL (PESOS)'])

print(f'Covarianza: {covarianza:,.2f}')
print(f'Correlación de Pearson: {coef_correlacion:.4f}')
```

```python
# Definición de variables independiente y dependiente
v_indep = df_1['VALOR MATRICULA (PESOS)']
v_depen = df_1['VALOR ARANCEL (PESOS)']

# Modelo de regresión lineal
modelo_lineal = np.polyfit(v_indep, v_depen, 1)
pendiente, intercepto = modelo_lineal

print(f'Ecuación: y = {pendiente:.2f}x + {intercepto:,.2f}')
```

```python
# Calcular predicciones según el modelo de regresión lineal obtenido
predicciones_v_depen = np.polyval(modelo_lineal, v_indep)

# Cálculo del coeficiente de determinación R²
R2 = r2_score(v_depen, predicciones_v_depen)

print(f'R² = {R2:.4f}')
```

```python
# Gráfico de dispersión
fig, ax = plt.subplots(figsize=(8, 6))
dispersion = ax.scatter(v_indep, v_depen, alpha=0.5)

# Línea de tendencia
ax.plot(v_indep, pendiente * v_indep + intercepto, color='red', 
        label=f'y={pendiente:.2f}x+{intercepto:.0f}')

# Personalización
ax.set_title('Relación Matrícula vs Arancel - GENERAL')
ax.set_xlabel('Valor Matrícula (Pesos)')
ax.set_ylabel('Valor Arancel (Pesos)')
ax.legend()
plt.show()
```

### 1.2 Análisis por Tipo de Institución

```python
# Ver tipos de instituciones
tipos = df_1['TIPO DE INSTITUCION'].unique()
print('Tipos de instituciones:')
for tipo in tipos:
    print(f'  - {tipo}')
```

```python
# Análisis por cada tipo
resultados = []

for tipo in tipos:
    df_tipo = df_1[df_1['TIPO DE INSTITUCION'] == tipo]
    
    # Covarianza
    cov = df_tipo['VALOR MATRICULA (PESOS)'].cov(df_tipo['VALOR ARANCEL (PESOS)'])
    
    # Correlación
    corr = df_tipo['VALOR MATRICULA (PESOS)'].corr(df_tipo['VALOR ARANCEL (PESOS)'])
    
    # Modelo
    v_i = df_tipo['VALOR MATRICULA (PESOS)']
    v_d = df_tipo['VALOR ARANCEL (PESOS)']
    modelo = np.polyfit(v_i, v_d, 1)
    m, b = modelo
    
    # R²
    pred = np.polyval(modelo, v_i)
    r2 = r2_score(v_d, pred)
    
    resultados.append({
        'Tipo': tipo,
        'Correlación': corr,
        'R²': r2,
        'Pendiente': m,
        'n': len(df_tipo)
    })
    
    print(f'\n{tipo}:')
    print(f'  Correlación: {corr:.4f}')
    print(f'  R²: {r2:.4f}')

# Crear tabla resumen
tabla = pd.DataFrame(resultados)
tabla
```

```python
# Gráficos por tipo
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, tipo in enumerate(tipos):
    df_tipo = df_1[df_1['TIPO DE INSTITUCION'] == tipo]
    v_i = df_tipo['VALOR MATRICULA (PESOS)']
    v_d = df_tipo['VALOR ARANCEL (PESOS)']
    
    modelo = np.polyfit(v_i, v_d, 1)
    m, b = modelo
    
    axes[idx].scatter(v_i, v_d, alpha=0.5)
    axes[idx].plot(v_i, m * v_i + b, 'r-', label=f'y={m:.2f}x+{b:.0f}')
    axes[idx].set_title(tipo)
    axes[idx].set_xlabel('Matrícula')
    axes[idx].set_ylabel('Arancel')
    axes[idx].legend()

plt.tight_layout()
plt.show()
```

```python
# Análisis de discrepancia
print('ANÁLISIS DE DISCREPANCIA')

if prob_educacion_femenino > prob_educacion_masculino:
    ratio = prob_educacion_femenino / prob_educacion_masculino
    print(f'\nSÍ existe discrepancia aparente:')
    print(f'Las mujeres tienen {ratio:.2f}x más probabilidad de estudiar')
    print(f'Educación que los hombres en la región del Maule 2021.')
    print(f'\n¿Cómo se explica?')
    print('• Históricamente, la carrera de Educación ha tenido mayor')
    print('  participación femenina')
    print('• Roles de género tradicionales asocian la enseñanza con')
    print('  características "femeninas"')
    print('• Mayor interés vocacional de las mujeres en áreas sociales')
    print('  y de cuidado')
else:
    ratio = prob_educacion_masculino / prob_educacion_femenino
    print(f'\nSÍ existe discrepancia aparente:')
    print(f'Los hombres tienen {ratio:.2f}x más probabilidad de estudiar')
    print(f'Educación que las mujeres en la región del Maule 2021.')
```
