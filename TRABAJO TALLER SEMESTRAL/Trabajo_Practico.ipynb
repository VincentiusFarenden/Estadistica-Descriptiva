**Tabla de frecuencias Variable EDAD**

```python
import numpy as np
import pandas as pd


EDAD = df_1['EDAD']  # Usando exactamente el nombre de la columna

# Paso 1: Rango
R = EDAD.max() - EDAD.min()

# Paso 2: Número de clases (Sturges)
n = len(EDAD)
k = int(1 + 3.322 * np.log10(n))

# Paso 3: Amplitud de clase
A = int(np.ceil(R / k))

print("Rango (R):", R)
print("Número de clases (k):", k)
print("Amplitud de clase (A):", A)

# Paso 4: Construcción de intervalos
bins = np.arange(EDAD.min(), EDAD.max() + A, A)
intervalos_EDAD = pd.cut(EDAD, bins=bins)

# Paso 5: Tabla de frecuencias
frecuencias_EDAD = intervalos_EDAD.value_counts().sort_index()

tabla_frecuencias_EDAD = pd.DataFrame({
    'Intervalo': frecuencias_EDAD.index.astype(str),
    'Frec. absoluta fi': frecuencias_EDAD.values,
    'Frec. relativa hi (%)': (frecuencias_EDAD.values / n) * 100,
    'Frec. abs. acumulada Fi': np.cumsum(frecuencias_EDAD.values),
    'Frec. rel. acumulada Hi (%)': np.cumsum(frecuencias_EDAD.values) / n * 100
})

# Redondear decimales a 1
tabla_frecuencias_EDAD = tabla_frecuencias_EDAD.round(1)

tabla_frecuencias_EDAD
```

```python
# Covarianza
covarianza = df_1['VALOR MATRICULA (PESOS)'].cov(df_1['VALOR ARANCEL (PESOS)'])

# Coeficiente de correlación lineal de Pearson
coef_correlacion = df_1['VALOR MATRICULA (PESOS)'].corr(df_1['VALOR ARANCEL (PESOS)'])

print(f'Covarianza: {covarianza:,.2f}')
print(f'Correlación de Pearson: {coef_correlacion:.4f}')

print('\nINTERPRETACIÓN EN CONTEXTO:')

print(f'\n• Covarianza = {covarianza:,.2f}')
if covarianza > 0:
    print('  → La covarianza positiva indica que cuando el valor de la')
    print('    matrícula aumenta, el valor del arancel también tiende a aumentar.')
else:
    print('  → La covarianza negativa indica una relación inversa.')
print('  Sin embargo, su magnitud es difícil de interpretar por sí sola.')
print(f'\n• Correlación de Pearson = {coef_correlacion:.4f}')
if abs(coef_correlacion) >= 0.8:
    fuerza = 'MUY FUERTE'
elif abs(coef_correlacion) >= 0.6:
    fuerza = 'FUERTE'
elif abs(coef_correlacion) >= 0.4:
    fuerza = 'MODERADA'
else:
    fuerza = 'DÉBIL'
print(f'  → Existe una correlación {fuerza} entre matrícula y arancel.')
print(f'  → Esto significa que el valor de la matrícula es un buen predictor')
print(f'    del valor del arancel en las instituciones de educación superior.')
```

```python
# Definición de variables independiente y dependiente
v_indep = df_1['VALOR MATRICULA (PESOS)']
v_depen = df_1['VALOR ARANCEL (PESOS)']

# Modelo de regresión lineal
modelo_lineal = np.polyfit(v_indep, v_depen, 1)
pendiente, intercepto = modelo_lineal

print(f'Ecuación: y = {pendiente:.2f}x + {intercepto:,.2f}')
```

```python
# Calcular predicciones según el modelo de regresión lineal obtenido
predicciones_v_depen = np.polyval(modelo_lineal, v_indep)

# Cálculo del coeficiente de determinación R²
R2 = r2_score(v_depen, predicciones_v_depen)

print(f'R² = {R2:.4f}')
```

**PREGUNTA 3: ¿Cuál es la probabilidad de seleccionar de los estudiantes hombres de la región a uno del área de educación? ¿Cuál es la probabilidad de seleccionar de las estudiantes mujeres de la región a una del área de educación? ¿Hay una discrepancia aparente? ¿Cómo se explica?**

```python
# Frecuencia absoluta: Género vs Área de Conocimiento
Tabla_genero_area = pd.crosstab(df_1['GENERO'], df_1['AREA CONOCIMIENTO'], margins=True)

print('Tabla de Frecuencias: Género vs Área de Conocimiento')
print(Tabla_genero_area)
```

```python
# Probabilidad condicional por filas (% dentro de cada género)
prob_area_dado_genero = pd.crosstab(df_1['GENERO'], df_1['AREA CONOCIMIENTO'],
                                     normalize='index', margins=True) * 100

print('Probabilidad Condicional: P(Área | Género) en %')
print(prob_area_dado_genero.round(2))
```

```python
# Extraer probabilidades específicas
if 'Educacion' in prob_area_dado_genero.columns:
    prob_educacion_masculino = prob_area_dado_genero.loc['Masculino', 'Educacion']
    prob_educacion_femenino = prob_area_dado_genero.loc['Femenino', 'Educacion']
else:
    print('NOTA: La columna "Educacion" no existe. Verificar nombre exacto en la base de datos.')
    print('Áreas disponibles:', prob_area_dado_genero.columns.tolist())
    # Intentar con nombre alternativo
    area_educacion = [col for col in prob_area_dado_genero.columns if 'educ' in col.lower()]
    if area_educacion:
        col_educacion = area_educacion[0]
        prob_educacion_masculino = prob_area_dado_genero.loc['Masculino', col_educacion]
        prob_educacion_femenino = prob_area_dado_genero.loc['Femenino', col_educacion]
        print(f'\nUsando columna: {col_educacion}')

print('PROBABILIDADES DE SELECCIONAR ESTUDIANTE DE EDUCACIÓN')
print(f'P(Educación | Hombre) = {prob_educacion_masculino:.2f}%')
print(f'P(Educación | Mujer) = {prob_educacion_femenino:.2f}%')
print(f'Diferencia: {abs(prob_educacion_femenino - prob_educacion_masculino):.2f} puntos porcentuales')
```

```python
# Visualización comparativa
fig, ax = plt.subplots(figsize=(8, 6))

generos = ['Hombres', 'Mujeres']
probabilidades = [prob_educacion_masculino, prob_educacion_femenino]

bars = ax.bar(generos, probabilidades, color=['#3498db', '#e74c3c'])

# TÍTULO CORREGIDO - Debe reflejar que es probabilidad CONDICIONAL
ax.set_title('Probabilidad Condicional: P(Educación | Género)\nRegión del Maule 2021', 
             fontsize=13, fontweight='bold')
ax.set_xlabel('Género', fontsize=11)
ax.set_ylabel('Probabilidad Condicional (%)', fontsize=11)
ax.set_ylim(0, max(probabilidades) * 1.2)
ax.grid(axis='y', alpha=0.3)

# Agregar valores
for bar, prob in zip(bars, probabilidades):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{prob:.2f}%', ha='center', fontweight='bold', fontsize=12)

# Agregar nota explicativa
ax.text(0.5, -0.15, 
        'P(Educación|Hombre) = % de hombres que estudian Educación\nP(Educación|Mujer) = % de mujeres que estudian Educación',
        ha='center', transform=ax.transAxes, fontsize=9, style='italic')

plt.tight_layout()
plt.show()
```

```python
# Análisis de discrepancia
print('ANÁLISIS DE DISCREPANCIA')

if prob_educacion_femenino > prob_educacion_masculino:
    ratio = prob_educacion_femenino / prob_educacion_masculino
    print(f'\nSÍ existe discrepancia aparente:')
    print(f'Las mujeres tienen {ratio:.2f}x más probabilidad de estudiar')
    print(f'Educación que los hombres en la región del Maule 2021.')
    print(f'\n¿Cómo se explica?')
    print('• Históricamente, la carrera de Educación ha tenido mayor')
    print('  participación femenina')
    print('• Roles de género tradicionales asocian la enseñanza con')
    print('  características "femeninas"')
    print('• Mayor interés vocacional de las mujeres en áreas sociales')
    print('  y de cuidado')
else:
    ratio = prob_educacion_masculino / prob_educacion_femenino
    print(f'\nSÍ existe discrepancia aparente:')
    print(f'Los hombres tienen {ratio:.2f}x más probabilidad de estudiar')
    print(f'Educación que las mujeres en la región del Maule 2021.')
```

```python
# Tabla resumen final
resumen = pd.DataFrame({
    'Género': ['Masculino', 'Femenino'],
    'Total estudiantes': [Tabla_genero_area.loc['Masculino', 'All'],
                          Tabla_genero_area.loc['Femenino', 'All']],
    'Estudiantes Educación': [Tabla_genero_area.loc['Masculino', col_educacion] if 'col_educacion' in locals() else Tabla_genero_area.loc['Masculino', 'Educacion'],
                               Tabla_genero_area.loc['Femenino', col_educacion] if 'col_educacion' in locals() else Tabla_genero_area.loc['Femenino', 'Educacion']],
    'P(Educación|Género) %': [prob_educacion_masculino, prob_educacion_femenino]
})

print('\nTABLA RESUMEN:')
print(resumen.to_string(index=False))
```

Esta discrepancia se explica por factores socioculturales:

4. **Factores económicos**: Las carreras de Educación tienen menor remuneración y se perciben como más compatibles con responsabilidades familiares.
5. **Modelos a seguir**: La predominancia femenina en docencia refuerza la tendencia.

**Conclusión:** La discrepancia NO es biológica sino resultado de construcciones sociales, segregación del mercado laboral y socialización diferencial por género. Este patrón se observa consistentemente en Chile y refleja desigualdades sistémicas en las elecciones educativas según género.
